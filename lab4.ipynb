```python
import matplotlib.pyplot as plt
import numpy as np

# var9
# 8.3 x1 + 8.2 x2 = 134
# –10.2 x1 + 15.2 x2 = 102
# 14.5 x1 – 7.3 x2 = 141

coef_eq1 = [8.3, 8.2]
const_eq1 = 134

coef_eq2 = [-10.2, 15.2]
const_eq2 = 102

coef_eq3 = [14.5, -7.3]
const_eq3 = 141

x_values = np.linspace(-10, 20, 100)
eq_1 = (const_eq1 - coef_eq1[0]*x_values)/coef_eq1[1]
eq_2 = (const_eq2 - coef_eq2[0]*x_values)/coef_eq2[1]
eq_3 = (const_eq3 - coef_eq3[0]*x_values)/coef_eq3[1]

fig, (ax1, ax2, ax3) = plt.subplots(nrows=3, ncols=1, figsize=(6, 12))


ax1.plot(x_values, eq_1, color='blue', linestyle='--')
ax1.set_ylim(-20, 20)
ax1.set_xlabel('x1')
ax1.set_ylabel('x2')
ax1.set_title('8.3 x1 + 8.2 x2 = 134')


ax2.plot(x_values, eq_2, color='orange', linestyle='-.')
ax2.set_ylim(-20, 20)
ax2.set_xlabel('x1')
ax2.set_ylabel('x2')
ax2.set_title(' –10.2 x1 + 15.2 x2 = 102')

ax3.plot(x_values, eq_3, color='green', linestyle=':')
ax3.set_ylim(-20, 20)
ax3.set_xlabel('x1')
ax3.set_ylabel('x2')
ax3.set_title('14.5 x1 – 7.3 x2 = 141')

plt.tight_layout()
plt.show()
```

```python
coeffs = [coef_eq1, coef_eq2, coef_eq3]
consts = [const_eq1, const_eq2, const_eq3]
points = []
for i in range(3):
    for j in range(i+1, 3):
        A = np.array([coeffs[i], coeffs[j]])
        b = np.array([consts[i], consts[j]])
        point = np.linalg.solve(A, b)
        points.append(point)


fig, ax = plt.subplots(figsize=(9, 16), dpi=100)

ax.plot(x_values, eq_1, color='blue', linestyle='--', label='8.3 x1 + 8.2 x2 = 134')
ax.plot(x_values, eq_2, color='orange', linestyle='-.', label=' –10.2 x1 + 15.2 x2 = 102')
ax.plot(x_values, eq_3, color='green', linestyle=':', label='14.5 x1 – 7.3 x2 = 141')
ax.legend()
ax.grid(color='gray', linestyle='--', linewidth=0.5)

ax.set_xlabel('x1')
ax.set_ylabel('x2')
ax.set_title('Intersection of Three Lines')

# Set the axis limits
ax.set_xlim(-20, 20)
ax.set_ylim(-20, 20)

ax.fill([p[0] for p in points], [p[1] for p in points], color='gray', alpha=0.5)

# Add labels for the intersection points
for i, point in enumerate(points):
    label = f'P{i+1} ({point[0]:.2f}, {point[1]:.2f})'
    ax.scatter([p[0] for p in points], [p[1] for p in points], c='red')
    ax.annotate(label, point, xytext=(10, 10), textcoords='offset points', fontsize=8, color='red')

# Display the plot
plt.show()


```
